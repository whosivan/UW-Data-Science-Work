### This is the Poisson part 

```python
# this library has a bunch of functions related to the binomial distribution 
from scipy.stats import poisson

# the range of data to plot the pdf (or pmf)
x=np.arange(20)

# some data sets binom.pmf(x,n,p) #x is our input range - it should cover the entire pdf/pmf 
y1=poisson.pmf(x,1.0)
y2=poisson.pmf(x,4.0)
y3=poisson.pmf(x,10.0)


# notice how to add legened! 
plt.figure(figsize=(4,8))
plt.subplot(211)
plt.plot(x,y1,label='v=1')
plt.plot(x,y2,label='v=4')
plt.plot(x,y3,label='v=10')
plt.legend()

# do the CDFs

y1=poisson.cdf(x,1.0)
y2=poisson.cdf(x,4.0)
y3=poisson.cdf(x,10.0)
plt.subplot(212)
plt.plot(x,y1,label='v=1')
plt.plot(x,y2,label='v=4')
plt.plot(x,y3,label='v=10')
plt.legend(loc=4)

```

### This is the normal distribution part 

* We didn't do this in class but if we had, this is what we would have done :O 

```python
# this library has a bunch of functions related to the binomial distribution 
from scipy.stats import norm

# the range of data to plot the pdf (or pmf)
x=np.arange(-10,10,0.01)

# 0 0 0 - 2 / .2 1 5 .5 
# some data sets binom.pmf(x,n,p) #x is our input range - it should cover the entire pdf/pmf 
y1=norm.pdf(x,0,0.2)
y2=norm.pdf(x,0,1.0)
y3=norm.pdf(x,0,5.0)
y4=norm.pdf(x,-2.0,0.5)

# notice how to add legened! 
plt.figure(figsize=(4,8))
plt.subplot(211)
plt.plot(x,y1,label='0,0.2 ')
plt.plot(x,y2,label='0,1.0')
plt.plot(x,y3,label='0,5.0 ')
plt.plot(x,y4,label='-2.0,0.5 ')

plt.legend()

# do the CDFs
y1=norm.cdf(x,0,0.2)
y2=norm.cdf(x,0,1.0)
y3=norm.cdf(x,0,5.0)
y4=norm.cdf(x,-2.0,0.5)

plt.subplot(212)
plt.plot(x,y1,label='0,0.2 ')
plt.plot(x,y2,label='0,1.0')
plt.plot(x,y3,label='0,5.0 ')
plt.plot(x,y4,label='-2.0,0.5 ')
```
